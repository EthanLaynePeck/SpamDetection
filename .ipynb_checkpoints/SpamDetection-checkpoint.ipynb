# 1. Data Import 

```python
import zipfile
import pandas as pd

zip_file_path = "data/smsspamcollection.zip"  
output_folder = "data/"  

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(output_folder)

    
data_path = "data/SMSSpamCollection"  
data = pd.read_csv(data_path, sep='\t', header=None, names=['label', 'message'])

data.head()
```

```python
print(data.isnull().sum())
print()


data.describe()
```

# 2. Exploration

```python
# What are the duplicated texts. There are only 5169 unique texts but 5572 total

duplicates = data[data[['message']].duplicated(keep = 'first')].copy()
duplicates
```

```python
spam_lost = round(duplicates[['label']].value_counts()[1]/data[['label']].value_counts()[1]*100, 1)
ham_lost = round(duplicates[['label']].value_counts()[0]/data[['label']].value_counts()[0]*100, 1)

print("If duplicates are removed:")
print(f"Spam Message Lost: {spam_lost}% ({duplicates[['label']].value_counts()[1]} records)")
print(f"Ham Message Lost: {ham_lost}% ({duplicates[['label']].value_counts()[0]} records)")
```

Lets take out the duplicates. If performance is bad we will put them back in. 

```python
data = data.drop_duplicates().copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='label')
plt.title("Spam vs. Ham Message Distribution")
plt.xlabel("Message Label")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud


def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=800, background_color="white", min_font_size=10).generate(text)
    plt.figure(figsize=(4, 4), facecolor=None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.title(title)
    plt.show()

# Generate word clouds for spam and ham messages
spam_text = " ".join(data[data['label'] == 'spam']['message'])
ham_text = " ".join(data[data['label'] == 'ham']['message'])

generate_wordcloud(spam_text, "Spam Messages Word Cloud")
generate_wordcloud(ham_text, "Ham Messages Word Cloud")
```

# 3. Cleaning Data

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# nltk.download('punkt')
# nltk.download('stopwords')

# Stemming is setting words to their "root" (connected/connecting -> connect, etc.)
# Snowball is the Porter2 stemming algorithm.
# I should try lemming sometime. Supposed to be better than stemming
stemmer = SnowballStemmer("english")

# Remove stop words (excellent resource to read more about why to remove stop words: https://wisdomml.in/what-are-stopwords-in-nlp-and-why-we-should-remove-them/ )
stop_words = set(stopwords.words("english"))

def preprocess_message(message):
    # Remove special characters and convert to lowercase
    message = re.sub(r'\W', ' ', str(message)).lower()
    
    # Tokenize
    words = word_tokenize(message)
    
    # Remove stopwords and stem
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    
    return ' '.join(words)

data['message_clean'] = data['message'].apply(preprocess_message)
```

# 4. Modeling Data

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['message_clean'], data['label'], test_size=0.5, random_state=666)

# Extract features using bag-of-words model (just counting tokens and putting counts into a vector)
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train the classifier
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)

print()
```

# 5. Evaluation

```python
# Make predictions
y_pred = classifier.predict(X_test_vec)

# Evaluate the model
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred, pos_label="spam"))
print("Recall: ", recall_score(y_test, y_pred, pos_label="spam"))
print("F1 Score: ", f1_score(y_test, y_pred, pos_label="spam"))
```

```python
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Create a DataFrame with the true and predicted labels
results = pd.DataFrame({'true_label': y_test, 'predicted_label': y_pred})

# Merge message data to predictions.
results = results.merge(data, left_index = True, right_index = True)


# Drop unnecessary columns
results = results.drop(columns=['label', 'message_clean'])

# Filter the DataFrame to show only mislabeled texts
mislabeled = results[results['true_label'] != results['predicted_label']]

mislabeled
```
